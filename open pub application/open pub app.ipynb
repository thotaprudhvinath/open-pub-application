```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"E:\Data Base\open_pubs.csv",header = None)
df
```

```python
df.columns=['Fsa_id','Name','Address','Postcode','Easting','Northing','Latitude','Longitude','Local_authority']
df.head()
```

```python
df.info()
```

- from the above info, we can say that there is no null values.
- and the data frame has 51331 rows and 9 columns.

```python
df['Latitude'] = df['Latitude'].replace(['\\N'],np.NaN)
df['Longitude'] = df['Longitude'].replace(['\\N'],np.NaN)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.head()
```

```python
df.info()
```

- after droping null values, total rows become 50564 and columns remains same.
- Latitude and Longitude dtype is object but it should be float.

```python
df['Longitude']=df['Longitude'].astype(float)
df['Latitude']=df['Latitude'].astype(float)
```

```python
df.to_csv(r"E:\Data Base\clean_open_pubs.csv",index=False)
```
